```python
import pandas as pd

# Load sentiment dataset
sentiment = pd.read_csv("fear_greed_index.csv")

# Load trader dataset
trades = pd.read_csv("historical_data.csv")

print("Sentiment shape:", sentiment.shape)
print("Trades shape:", trades.shape)

sentiment.head()
```

```python
print(sentiment.isnull().sum())
print(trades.isnull().sum())

print("Duplicates in sentiment:", sentiment.duplicated().sum())
print("Duplicates in trades:", trades.duplicated().sum())
```

```python
sentiment["date"] = pd.to_datetime(sentiment["date"])
sentiment["day"] = sentiment["date"].dt.date
```

```python
trades["Timestamp IST"] = pd.to_datetime(trades["Timestamp IST"], format='%d-%m-%Y %H:%M')
trades["day"] = trades["Timestamp IST"].dt.date
```

```python
merged = trades.merge(sentiment, on="day", how="inner")

print("Merged shape:", merged.shape)
merged.head()
```

```python
daily_pnl = merged.groupby(["Account", "day", "classification"])["Closed PnL"].sum().reset_index()

daily_pnl.head()
```

```python
merged["win_trade"] = merged["Closed PnL"] > 0

win_rate = merged.groupby("classification")["win_trade"].mean()

print(win_rate)
```

```python
print("Sentiment day range:", sentiment['day'].min(), "to", sentiment['day'].max())
print("Trades day range:", trades['day'].min(), "to", trades['day'].max())
```

```python
trades_per_day = merged.groupby(["day", "classification"]).size().reset_index(name="num_trades")

trades_per_day.head()
```

```python
long_short = merged.groupby(["classification", "Side"]).size().unstack().fillna(0)

long_short["long_short_ratio"] = long_short["BUY"] / (long_short["SELL"] + 1)

print(long_short)
```

```python
import matplotlib.pyplot as plt

pnl_summary = daily_pnl.groupby("classification")["Closed PnL"].mean()

pnl_summary.plot(kind="bar")
plt.title("Average Daily PnL: Fear vs Greed")
plt.show()
```

```python
win_rate.plot(kind="bar")
plt.title("Win Rate: Fear vs Greed")
plt.show()
```

```python
merged.boxplot(column="Size USD", by="classification")
plt.title("Trade Size Distribution by Sentiment")
plt.suptitle("")
plt.show()
```

```python
activity = merged.groupby("Account").size()

median_trades = activity.median()

merged["activity_group"] = merged["Account"].apply(
    lambda x: "Frequent" if activity[x] > median_trades else "Infrequent"
)

activity_result = merged.groupby(["activity_group", "classification"])["Closed PnL"].mean()

print(activity_result)
```
